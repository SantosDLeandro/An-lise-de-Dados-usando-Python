```python
def truncar(Descrição):
    return Descrição[:4]
portal['Descrição'].apply(truncar).value_counts()#Exibi a coluna, retorne a funcao truncar e exiba os valores consolidados de empenhos por entidade
```

```python
#Criando o Grafico
cnpj_cpf_lancados = portal['Descrição'].apply(truncar).value_counts().head(10)
cnpj_cpf_lancados.plot(kind='bar', figsize=(11,5), grid = False, rot=0, color='green')

#Definido Títulos
plt.title('As 10 Primeiras entidades com maior numero de Empenhos')
plt.xlabel('Entidades') #Nomeando o eixo X
plt.ylabel('Quant de Empenhos lançados')#Nomeando o eixo Y
```

```python
#Cruzando informações de Tabelas invertidas em ref a anterior
cruztab_dv = pd.crosstab(portal['Descrição'], portal['Valor Pago'])
cruztab_dv.head()
```

```python
cruztab_dv['total'] = cruztab_dv.sum(axis=1)
cruztab_dv.head()
```

```python
top10_emp = cruztab_dv[cruztab_dv['total']>10000].sort_values('total', ascending = False)
top10_final = top10_emp.append(pd.DataFrame(top10_emp.sum(), columns=['total']).T, ignore_index=False)

sns.set(font_scale=1)
plt.figure(figsize=(18,9))
sns.heatmap(top10_final, annot=True)
```
